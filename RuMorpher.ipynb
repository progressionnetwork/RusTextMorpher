```python
# -*- coding: utf-8 -*-
from collections import Counter
from itertools import chain
import textwrap
import difflib
import regex
import random
from string import ascii_letters
from transliterate.utils import translit # !pip install transliterate==1.10.2
from langdetect import detect # !pip install langdetect
 
# Configurator
shuffle_uniqcode = True
shuffle_wnlines = True
shuffle_uppercase = True
shuffle_errors = True
shuffle_translit = True

# variable character set
char_a = ["a", 'а']
char_c = ["с", 'c']
char_C = ["С", 'C']
char_e = ["е", 'e']
char_E = ["Е", 'E']
char_ye = ["ё", 'e', 'е']
char_o = ["о", 'o']
char_O = ["О", 'O']
char_A = ['А', 'A']
char_minus = ["-", '--']
char_dot = [".", ';']
list_str = []

# variable with unicode chars
if shuffle_uniqcode:
    char_a = ["a", 'а', 'ā', 'ă']
    char_c = ["с", 'c', 'ç']
    char_C = ["С", 'C', 'Ć']
    char_e = ["е", 'e', 'ę']
    char_E = ["Е", 'E', 'Ē']
    char_ye = ["ё", 'ě', char_e, 'ē']
    char_o = ["о", 'o', 'ō', 'ŏ', 'ő']
    char_O = ["О", 'O', 'Ő', 'Ō', 'Ŏ']
    char_A = ['А', 'A', 'Ā', 'Ă']
    char_minus = ["-", '−', '–', '—', '--']
    char_dot = [".", ';', '․', '·', '·']


def insert_newlines(str, step=random.randrange(80, 100)):
    return '\n'.join(textwrap.wrap(str, step))

def similar(a, b):
    return difflib.SequenceMatcher(None, a, b).ratio()

def rndone():
    if random.randrange(0, 2) == 1:
        return True
    

str = """Суперхит-инструкция к простой жизни: сказать «нет», когда нужно, сделать из неудач мост к переменам или радоваться, с легкостью решая любые проблемы.
Нескучная философия: офисный работник, погрязший в рутине, забредает в кафе «Почему». Здесь он находит новый смысл жизни и осознает свои желания.
"""

# str = """It may be audacious even to attempt a definition of his greatness, but it is not so difficult to describe the gifts that enabled him to create imaginative visions of pathos and mirth that, whether read or witnessed in the theatre, fill the mind and linger there. He is a writer of great intellectual rapidity, perceptiveness, and poetic power. 
# """

def obscure_text(str):
    
    print('============ Original text: =================')
    print (str)
    print('=============================================')

    lng = detect(str)
    print ('Original Language detection: ', lng)

    str_m = str

    # shuffle text with translit
    if shuffle_translit:
        try:
            str_m = translit(str_m, lng, reversed=True)
        except:
            print("Unable to shuffle with translit")

    # randomly replace chars
    if lng == 'ru':
        if rndone():
            str_m = str_m.replace("а", random.choice(char_a))
        if rndone():
            str_m = str_m.replace("А", random.choice(char_A))
        if rndone():
            str_m = str_m.replace("о", random.choice(char_o))
        if rndone():
            str_m = str_m.replace("О", random.choice(char_O))
        if rndone():
            str_m = str_m.replace("с", random.choice(char_c))
        if rndone():
            str_m = str_m.replace("С", random.choice(char_C))
        if rndone():
            str_m = str_m.replace("е", random.choice(char_e))
        if rndone():
            str_m = str_m.replace("Е", random.choice(char_E))
        if rndone():
            str_m = str_m.replace("ё", random.choice(char_ye))
        if rndone():
            str_m = str_m.replace("р", "p")
        if rndone():
            str_m = str_m.replace("Р", "P")
        if rndone():
            str_m = str_m.replace("Н", "H")
        if rndone():
            str_m = str_m.replace("К", "K")
        if rndone():
            str_m = str_m.replace("у", "у")
        if rndone():
            str_m = str_m.replace("У", "Y")
        if rndone():
            str_m = str_m.replace("В", "B")
        if rndone():
            str_m = str_m.replace("х", "x")
        if rndone():
            str_m = str_m.replace("Х", "X")
        if rndone():
            str_m = str_m.replace("ъ", "ь")

    if lng == 'en':
        if rndone():
            str_m = str_m.replace("a", random.choice(char_a))
        if rndone():
            str_m = str_m.replace("A", random.choice(char_A))
        if rndone():
            str_m = str_m.replace("o", random.choice(char_o))
        if rndone():
            str_m = str_m.replace("O", random.choice(char_O))
        if rndone():
            str_m = str_m.replace("c", random.choice(char_c))
        if rndone():
            str_m = str_m.replace("C", random.choice(char_C))
        if rndone():
            str_m = str_m.replace("e", random.choice(char_e))
        if rndone():
            str_m = str_m.replace("E", random.choice(char_E))
        if rndone():
            str_m = str_m.replace("p", "р")
        if rndone():
            str_m = str_m.replace("P", "Р")
        if rndone():
            str_m = str_m.replace("H", "Н")
        if rndone():
            str_m = str_m.replace("K", "К")
        if rndone():
            str_m = str_m.replace("у", "у")
        if rndone():
            str_m = str_m.replace("Y", "У")
        if rndone():
            str_m = str_m.replace("B", "В")
        if rndone():
            str_m = str_m.replace("x", "х")
        if rndone():
            str_m = str_m.replace("X", "Х")
        if rndone():
            str_m = str_m.replace("t", "Т")

    if rndone():
        str_m = str_m.replace("-", random.choice(char_minus))
    if rndone():
        str_m = str_m.replace(".", random.choice(char_dot))
    if rndone():    
        if shuffle_wnlines:
            str_m = str_m.replace("\n", "\n\n", random.randrange(0, 2))

    # shuffle text with random uppercase chars
    if shuffle_uppercase:
        list_str = list(str_m)
        for idx, char in enumerate(list_str):
            dice = random.randint(1,6)
            if dice > 5:
                char = char.swapcase()
                list_str[idx] = char
        str_m = ''.join(list_str)

    # shuffle text with new lines
    if shuffle_wnlines:
        str_m = insert_newlines(str_m)

    # shuffle text with random erros
    if shuffle_errors:
        chars = [i for i,_ in enumerate(str_m) if not str_m.isspace()]
        dice = random.sample(chars, 3)
        letters =  iter(random.sample(ascii_letters, 3))
        list_str = list(str_m)
        for ind in dice:
            list_str[ind] = next(letters)
        str_m = ''.join(list_str)

    print('============ Obfuscated text: ===============')
    print (str_m)
    print('=============================================')

    lng = detect(str_m)
    print ('Obscured Language detection: ', lng)

    a = str
    b = str_m

    print("Similararity:", similar(a,b))

    print('=============================================')
    print('Diff details:\n')
    for i,s in enumerate(difflib.ndiff(a, b)):
        if s[0]==' ': continue
        elif s[0]=='-':
            print(u'Delete "{}" from position {}'.format(s[-1],i))
        elif s[0]=='+':
            print(u'Add "{}" to position {}'.format(s[-1],i))    

    print('=============================================')
            
    counts = Counter(chain.from_iterable(str))
    counts_m = Counter(chain.from_iterable(str_m))

    total = sum(counts.values())
    total_m = sum(counts_m.values())

    chars = set('АаВвСсХхaфϕУуРрЕеОоAoOcCeEpPHKyYBxXˌ\n')

    for character, count in counts.items():
        if character in chars:
            print('"{:<2}" - {:>6.2f}%'.format(repr(character)[1:-1], (count/total) * 100))
        
    print('=============================================')

    for character, count in counts_m.items():
        if character in chars:
            print('"{:<2}" - {:>6.2f}% '.format(repr(character)[1:-1], (count/total_m) * 100))

if __name__ == "__main__":
    obscure_text(str)
```
